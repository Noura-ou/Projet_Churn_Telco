Création de la variable nb_service_internet

```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("../data/silver.csv")
df.head()
```

```python

internet_services_columns = ['OnlineSecurity', 'OnlineBackup','DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies']
df['nb-service'] = df[internet_services_columns].apply(lambda x: x == 'Yes').sum(axis=1)
```
