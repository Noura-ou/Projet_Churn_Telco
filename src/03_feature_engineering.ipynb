Cr√©ation de la variable nb_service_internet

```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("../data/silver.csv")
df.head()
```

```python
def add_nb_service_column(df):
    services = ['PhoneService', 'OnlineSecurity', 'OnlineBackup', 
                'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies']
    df['nb-service'] = df[services].apply(lambda x: x.eq('Yes').sum(), axis=1)
    df = df.drop(columns=services)
    return df
```

```python
df = add_nb_service_column(df)
df.head()
```
